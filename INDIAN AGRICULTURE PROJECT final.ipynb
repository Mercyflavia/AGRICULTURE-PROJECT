```python
##Correlation Heatmap
#heatmap of the correlation matrix helps us understand the relationships between multiple variables.

# Correlation matrix
correlation_matrix = df[['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)', 
                         'SUGARCANE PRODUCTION (1000 tons)', 'OILSEEDS PRODUCTION (1000 tons)',
                         'SUNFLOWER PRODUCTION (1000 tons)', 'COTTON PRODUCTION (1000 tons)']].corr()

# Plotting the heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Top 7 Rice producing states (with observed=False to silence the warning)
top_rice_producers = df.groupby('State Name', observed=False)['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7)

# Bar chart for Rice production
plt.figure(figsize=(7, 5))  # Adjusted for better readability
top_rice_producers.plot(kind='bar', color='green')
plt.title('Top 7 Rice Producing States')
plt.xlabel('State')
plt.ylabel('Rice Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()

```

```python
# Top 5 Wheat producing states (with observed=False to silence the warning)
top_wheat_producers = df.groupby('State Name', observed=False)['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5)

# Bar chart for Wheat production
plt.figure(figsize=(7, 5))  # Adjusted for better readability
top_wheat_producers.plot(kind='bar', color='orange')
plt.title('Top 5 Wheat Producing States')
plt.xlabel('State')
plt.ylabel('Wheat Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()

# Pie chart for Wheat production percentage
plt.figure(figsize=(6, 6))  # Adjusted for better readability
top_wheat_producers.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Wheat Production Percentage Share of Top 5 States')
plt.ylabel('')
plt.show()

```

```python
# Group by State and sum the oilseed production (with observed=False to silence the warning)
top_oilseed_producers = df.groupby('State Name', observed=False)['OILSEEDS PRODUCTION (1000 tons)'].sum().reset_index()

# Sort by oilseed production and take top 5 states
top_oilseed_states = top_oilseed_producers.sort_values(by='OILSEEDS PRODUCTION (1000 tons)', ascending=False).head(5)

# Bar Plot for Oilseed Production by Top 5 States using Plotly
import plotly.express as px

fig = px.bar(top_oilseed_states, x='State Name', y='OILSEEDS PRODUCTION (1000 tons)',
             title='Oilseed Production by Top 5 States',
             labels={'OILSEEDS PRODUCTION (1000 tons)': 'Oilseed Production (1000 tons)', 'State Name': 'State'},
             color='State Name',  # Optional: Adds colors to each bar based on state
             text='OILSEEDS PRODUCTION (1000 tons)')  # Optional: Adds the value on top of the bars
fig.update_traces(texttemplate='%{text}', textposition='outside')  # Adjusting the position of the text
fig.show()
```

```python
# Group by State and sum the sunflower production (with observed=False to silence the warning)
top_sunflower_producers = df.groupby('State Name', observed=False)['SUNFLOWER PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7)

# Plotting the bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
top_sunflower_producers.plot(kind='bar', color='gold')
plt.title('Top 7 Sunflower Producing States')
plt.xlabel('State')
plt.ylabel('Sunflower Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()

```

```python


# Grouping by Year for both Rice and Wheat production
production_by_year = df.groupby('Year')[['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)']].sum()

# Plotting the line chart for both
plt.figure(figsize=(12, 6))
production_by_year.plot(kind='line')
plt.title('Rice vs Wheat Production Over the Last 50 Years')
plt.xlabel('Year')
plt.ylabel('Production (1000 tons)')
plt.grid(True)
plt.legend(['Rice Production', 'Wheat Production'])
plt.show()

# Optionally, if you want to remove the warning permanently, you could use observed=False:
# production_by_year = df.groupby('Year', observed=False)[['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)']].sum()
```

```python
# Top 5 Soybean producing states
top_soybean_producers = df.groupby('State Name')['SOYABEAN PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5)

# Yield Efficiency calculation (Production / Area)
df['SOYABEAN YIELD EFFICIENCY'] = df['SOYABEAN PRODUCTION (1000 tons)'] / df['SOYABEAN AREA (1000 ha)']

# Plotting the bar chart for Soybean production
plt.figure(figsize=(7, 4))
top_soybean_producers.plot(kind='bar', color='purple')
plt.title('Top 5 Soybean Producing States')
plt.xlabel('State')
plt.ylabel('Soybean Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()

# Plotting Yield Efficiency for Top 5 Soybean Producing States
# Filter data for the top soybean producers
top_soybean_data = df[df['State Name'].isin(top_soybean_producers.index)]

# Plotting Yield Efficiency
plt.figure(figsize=(12, 5))
sns.barplot(x='State Name', y='SOYABEAN YIELD EFFICIENCY', data=top_soybean_data, palette='viridis')
plt.title('Soybean Yield Efficiency in Top 5 Soybean Producing States')
plt.xlabel('State')
plt.ylabel('Soybean Yield Efficiency (Kg per ha)')
plt.xticks(rotation=45)
plt.show()
```
